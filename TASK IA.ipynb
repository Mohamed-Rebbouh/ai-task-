part 3 of question 3 ds:

```python
def min_max_hourly_wage(val):
    if pd.isnull(val):
        return np.nan
    elif ' /hour' in val:
        i=val.split('/hour')[0].replace('$',' ').strip()
        if '-' in i:
            mn=i.split('-')[0]
            mx=i.split('-')[1]
#Since there are several commas in salary. It will take it as a string. So it will throw an 
#error saying "unsupported operand type(s) for /: 'str' and 'int'" and will not be able to 
#perform operation. And, I used try catch block because there is € currency in salary also. So in order 
#to deal with this I preferred try catch block.
            try:
                mn = float(mn.replace(",", "").strip())
                mx = float(mx.replace(",", "").strip())
            except:
                return np.nan
            return mn, mx


def min_max_yearly_wage(val):
    if pd.isnull(val):
        return np.nan
    elif ' /year' in val:
        i=val.split('/year')[0].replace('$',' ').strip()
        if '-' in i:
            mn=i.split('-')[0]
            mx=i.split('-')[1]
#Since there are several commas in salary. It will take it as a string. So it will throw an 
#error saying "unsupported operand type(s) for /: 'str' and 'int'" and will not be able to 
#perform operation. And, I used try catch block because there is € currency in salary also. So in order 
#to deal with this I preferred try catch block.
            try:
                mn = float(mn.replace(",", "").strip())
                mx = float(mx.replace(",", "").strip())
            except:
                return np.nan
            return mn, mx
        
        
cdf = cdf.assign(yearly_salary_range=cdf['salary'].apply(min_max_yearly_wage),
                   hourly_salary_range=cdf['salary'].apply(min_max_hourly_wage)) 
```

```python
cdf = cdf.assign(
    median_yearly_salary = cdf['yearly_salary_range'].apply(
        lambda r: (r[0] + r[1]) / 2 if pd.notnull(r) else r
    ),
    median_hourly_salary = cdf['hourly_salary_range'].apply(
        lambda r: (r[0] + r[1]) / 2 if pd.notnull(r) else r
    )
)
```

```python
cdf.groupby(by=['organization'], as_index=False)['median_hourly_salary'].mean().sort_values(by='median_hourly_salary',ascending=False)
cdf.groupby(by=['organization'], as_index=False)['median_yearly_salary'].mean().sort_values(by='median_yearly_salary',ascending=False)
```

```python
cdf.head()
```

```python
cdf=cdf.drop(['yearly_salary_range','hourly_salary_range'],axis=1)
#
cdf['median_hourly_salary']=cdf['median_hourly_salary']*8*26
cdf['median_yearly_salary']=cdf['median_yearly_salary']/12      
#
cdf['median_yearly_salary'].fillna(cdf['median_hourly_salary'], inplace=True)
#
cdf['salary']=cdf['median_hourly_salary']
cdf.head()
```

```python
cdf=cdf.sort_values(by=["salary"],ascending=False)
cdf.head(10)
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

```

```python
sdf=cdf.dropna(subset=['salary'])
sdf.info()
```

```python
le = LabelEncoder()
sdf['job_type'] = le.fit_transform(sdf['job_type'])
sdf['job_description'] = le.fit_transform(sdf['job_description'])
```

```python
plt.scatter(sdf.job_type, sdf.salary,  color='blue')
plt.xlabel("typpe")
plt.ylabel("salry")
plt.show()
```

```python
sdf.head()
```

```python
sdf=sdf.drop(['median_yearly_salary','median_hourly_salary'],axis=1)
```

```python
sdf.info()
```

```python
sdf["job_class"]=""
sdf["job_class"][sdf['salary']>=20000000.00 ]="high"
sdf["job_class"][sdf['salary']<5000000]="low"
sdf["job_class"][sdf['job_class']==""]="medio"
salary={'high': 20000000.00,
        'medio': 8000000,
        'low':3000}








sdf["job_class"].value_counts()
```

```python
sdf["salary"][sdf["job_class"]=="high"]=salary['high']
sdf["salary"][sdf["job_class"]=="low"]=salary['low']
sdf["salary"][sdf["job_class"]=="medio"]=salary['medio']
sdf["salary"].value_counts()
```

```python
X=sdf[['job_description','job_type']]
X_train, X_test, y_train, y_test = train_test_split(X,sdf['salary'], test_size=0.2)
```

```python
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
acc=accuracy_score(y_pred,y_test)
print("accuracy test=",acc)
```
